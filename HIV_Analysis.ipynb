This is not as relaible as the percentage of each community, because it does not take into account how many poeple were tested for each group.


```python
AIDS_Education = df2[df2['AIDS education'] == 'YES']
AIDS_YES = AIDS_Education[['AIDS education', 'Result']]
AIDS_YES.head()
```

```python
AID_EDUCATION_NEGATIVE = AIDS_YES['Result'].value_counts()['NEGATIVE']
AID_EDUCATION_NEGATIVE
```

```python
AID_EDUCATION_POSITIVE = AIDS_YES['Result'].value_counts()['POSITIVE']
AID_EDUCATION_POSITIVE
```

```python
AID_EDUCATION_COMPARISON = ['AID_EDUCATION_NEGATIVE', 'AID_EDUCATION_POSITIVE']
Value = [178,105]
plt.bar(AID_EDUCATION_COMPARISON, Value)

plt.title('Have AIDS Education and Result')
plt.ylabel('Number of People')
```

**Analysis :** People who have acess to AIDS education are 1.69 times more likely to test negative for HIV than test positive.
